```python
import numpy as np

X = np.asarray([[2,3,3,4,5,7],[2,4,5,5,6,8]])
print(X)
```

```python
#1: 
cov = X.T @ X
(eigenvalues_xtx, eigenvector_xtx) = np.linalg.eig(cov)

# print(eigenvector_xtx)
print(eigenvalues_xtx)
```

```python
#2: Verify
XXt = X @ X.T
print(XXt)
print(np.linalg.eig(XXt)[0])
```

```python
#4
subtracted = X.T - np.mean(X, axis=1)
normalized = X.T - np.std(X,axis=1)
cov = np.cov(normalized.T @ normalized)
eigval, eigvec = np.linalg.eig(cov)
print(eigvec)
```

```python
#take top eigval
top = eigvec[1]
projected_data = X.T @ top
print(projected_data)
```

```python
Y = [1,1,0,1,0,0,1,0,1,1]
Theta1 = (.5,.5,.5) 
Theta2 = (.4,.6,.7)
thetas = []
thetas.append(Theta1)
thetas.append(Theta2)
for theta in thetas:
    print("EM for Theta_0 = ", theta)
    for i in range(20):
        #Expectation
        print("Iteration:", i)
        mu = []
        for y in Y:
            u = (theta[0]*theta[1]**(y)*(1-theta[1])**(1-y))/(theta[0]*theta[1]**(y)*(1-theta[1])**(1-y) + (1-theta[0])*theta[2]**(y)*(1-theta[2])**(1-y))
            mu.append(u)
        print("mu=",mu)
        
        ##Maximization
        pi = np.sum(mu)/10
        print("pi=", pi)
        p = 0
        for i in range(len(mu)):
            p+= Y[i]*mu[i]
        p = p/(np.sum(mu))
        print("p=",p)
        q=0
        denom = 0
        for i in range(len(mu)):
            q+= Y[i] * (1-mu[i])
            denom += (1-mu[i])
        q = q/denom
        print("q=",q)
        
```
